## ‚öôÔ∏è Configuraci√≥n de Seaborn

```python
# Configurar seaborn para que sea consistente con matplotlib
try:
    import seaborn as sns
    
    # Configuraci√≥n personalizada de seaborn que respeta matplotlib
    sns.set_theme(
        context="notebook",   # Tama√±o base
        style="whitegrid",    # Estilo con grid
        palette=[             # Misma paleta de colores tierra
            '#D2691E',  # Chocolate/naranja tierra
            '#A0522D',  # Marr√≥n siena  
            '#CD853F',  # Dorado tierra
            '#8B4513',  # Marr√≥n silla de montar
            '#DAA520',  # Vara de oro
            '#B8860B'   # Vara de oro oscura
        ],
        rc={
            # Configuraciones que coincidan con matplotlib
            'axes.facecolor': '#fefefe',      # Mismo fondo
            'axes.edgecolor': '#8B7355',      # Mismo borde
            'grid.alpha': 0.3,                # Misma transparencia grid
            'lines.linewidth': 0.75,          # Mismo grosor l√≠neas
            'font.family': ['Montserrat', 'Arial', 'DejaVu Sans'],  # Mismas fuentes
            'figure.facecolor': 'white',      # Fondo figura
            'axes.spines.top': False,         # Sin espina superior
            'axes.spines.right': False,       # Sin espina derecha
        }
    )
    
    print("‚úÖ Seaborn configurado con mismos par√°metros que matplotlib")
    print("üé® Paleta tierra aplicada a seaborn")
    
except ImportError:
    print("‚ö†Ô∏è Seaborn no disponible - usando solo matplotlib")
except Exception as e:
    print(f"‚ö†Ô∏è Error configurando seaborn: {e}")
    print("üìä Continuando con matplotlib")
```
